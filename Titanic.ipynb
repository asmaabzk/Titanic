```python
df = df.drop(['Cabin', 'Ticket','PassengerId', 'Name'], axis= 1)
```

```python
df
```

```python
fq_Embarked = df.Embarked.dropna().mode()[0]
```

```python
fq_Embarked
```

```python
df['Embarked'] = df['Embarked'].fillna(fq_Embarked)
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].dropna().median(), inplace= True)
df['Fare'].fillna(df['Fare'].dropna().median(), inplace= True)
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
df['Sex'] = LabelEncoder().fit_transform(df['Sex'])
```

```python
df.Sex
```

```python
df['Embarked'] = LabelEncoder().fit_transform(df['Embarked'])
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
X_df = df.drop("Survived", axis=1)
y_df = df["Survived"]
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, test_size=0.3, random_state=0)
```

```python
from sklearn.neural_network import MLPClassifier
clf = MLPClassifier(hidden_layer_sizes=(50)).fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
```

```python
accuracy
```

```python
clf2=MLPClassifier(solver='adam',hidden_layer_sizes=(174,38),activation='logistic', random_state=1).fit(X_train,y_train)
print("Accuracy: ",clf2.score(X_test,y_test))
```

```python
clf2=MLPClassifier(hidden_layer_sizes=(150,50), solver='lbfgs',activation='logistic', alpha=0.1).fit(X_train,y_train)
print("Accuracy",clf2.score(X_test,y_test))
```
